```python
from datetime import timedelta, datetime

#df['readable_timestamp'] = pd.to_datetime(df['context_timestamp'], unit='s')
#df['readable_timestamp'] = df['readable_timestamp'].dt.strftime(date_format="%Y-%m-%d")
df['readable_timestamp'] = df['context_timestamp'].apply(lambda x: datetime.fromtimestamp(x).date())
df = df.sort_values(by="readable_timestamp")
df.head()
```

```python
alldays = df['readable_timestamp'].unique()
#alldays.datetime.date()
alldays
```

```python
# In the past 3 days, we calculate:
df0 = pd.DataFrame({})
for end_day in alldays:
    start_day = end_day - timedelta(days=3)
    last3days = df[(df['readable_timestamp'] <= end_day) & (df['readable_timestamp'] >= start_day)]
    #total_clicks = last3days['is_trade'].count()
    #total_trades = last3days.groupby('item_id')['is_trade'].sum()
    # 历史转化率 =  历史交易数量/历史总点击数量, which is the mean()
    df_tradeRate = last3days.groupby('item_id').mean()[['is_trade']]
    df_tradeRate['readable_timestamp'] = pd.Series([end_day for i in range(df_tradeRate.shape[0])], index=df_tradeRate.index)
    df0 = pd.concat([df0, df_tradeRate], axis = 0)

df0['item_id'] = df0.index
df_tradeRate = df0.rename(columns={'is_trade':'trade_rate'})
df_tradeRate.shape
```

```python
df_tradeRate.head()
```

```python
df = df.merge(df_tradeRate, on=['item_id', 'readable_timestamp'], how='left')
df.head()
```
